````
import matplotlib.pyplot as plt
import numpy as np
````

````

h= 66.626e-34       #constante de plank
c = 3.0e+8          # velocidad de la luz en micrometros
kB = 11.38e-23      # contante de Bolztmann
````

formula
$$ B(\lambda ,T)={\frac {2hc^{2}}{\lambda ^{5}}}{\frac {1}{e^{\frac {hc}{\lambda k_{\mathrm {B} }T}}-1}} $$
```
# Tiene formato de c√≥digo

```



````
# definir el rango de la longitud de onda en micrometros
wave = np.arange(1e-9, 3e-6, 1e-9)

# definir una funcion para calcular la formula de Plank
def planck(wavelength, T):
    a = 2.0*h*c**2
    b = h*c/(wavelength*kB*T)
    sbs = a/ ( (wavelength**5) * (np.exp(b) - 1.0) )
    return sbs
    
````

````
 #aqui en cada plot individual llamamos la funcion plank definida arriba por Temperatura T = 2000, 3000 ...

plt.plot(wave*1.0e9, planck(wave, 5000), 'r')
plt.plot(wave*1.0e9, planck(wave, 6000), 'g')
plt.plot(wave*1.0e9, planck(wave, 7000), 'y')
plt.plot(wave*1.0e9, planck(wave, 8000), 'm')
#plt.axis([1,3000,1000,2.0e14])
#plt.xscale('log')
#plt.yscale('log')
plt.ylabel("spectral radiation flux")
plt.xlabel("wavelength (um)")
plt.show()
````
